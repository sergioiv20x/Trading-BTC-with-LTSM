```python
# Read all the CSV files that start with 'output_got.....' to produce a list containing: 
# filename, min date of collected Tweet in file, max. Date of collected Tweet in file, count of colelcted tweets in file.
# based on the code on https://github.com/ekapope/Combine-CSV-files-in-the-folder/blob/master/Combine_CSVs.py
import os
import pandas as pd
directory = os.getcwd() # obtain initial directory
lista=[]
for file in os.listdir(directory):
    filename = os.fsdecode(file)
    if filename.startswith("output_got"):
        #print(os.path.join(filename))
        df=pd.read_csv(filename, sep=';', error_bad_lines=False)
        ma=df['date'].max()
        mi=df['date'].min()
        lista.append([filename,mi,ma,len(df)])
    else:
        continue
```

```python
#view the produced list
lista
```

```python
#find all csv files in the folder
#use glob pattern matching -> extension = 'csv'
#save result in list -> all_filenames
import os
import glob
import pandas as pd
os.chdir(directory)# set working directory
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
combined_csv = pd.concat([pd.read_csv(f, sep=';', error_bad_lines=False) for f in all_filenames ],sort=False) #combine all files in the list
#combined_csv.to_csv("combined_csv_2017_01.csv", index=True,sep=';') #export to csv
print('creating combined CSV')
```

```python
# Here we check column number of entries and duplicates:
A=len(combined_csv)
df=combined_csv[combined_csv["Unnamed: 10"].isnull()] # ignoring collected entries with more than 10 columns
df = df.drop_duplicates(subset='id', keep='first')  # Delete duplicated entries based on the Id column
print('Deleted {} duplicates from {} collected'.format(A-len(df),A))
del df['username']
del df['retweets']
del df['favorites']
del df['geo']
del df['mentions']
del df['hashtags']
del df['id']
del df['permalink']
del df['Unnamed: 10']
```

```python
df.date=pd.to_datetime(df.date) #date format for Pandas Dataframe
```

```python
df=df.groupby([pd.Grouper(key='date',freq='H')]).size().reset_index(name='count') # Count of tweets per hour
```
